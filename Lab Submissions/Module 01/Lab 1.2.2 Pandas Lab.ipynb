# 5. Visualization

> Indented block



#### 5.A Plot Histogram of Age 

````python
# A:
plt.hist(df['age'], bins = 20)
plt.title("Histogram of Age")
plt.xlabel("Age of Employee")
plt.ylabel('Frequency')
plt.show()
````

#### 5.B BoxPlot of Age

````python
# A:
plt.boxplot(df['age'])
plt.title("Boxplot of Age")
plt.ylabel('Age of Employee')
plt.show()
````

#### 5.C Scatterplot of age and length_of_service

````python
# A:
plt.scatter(df['age'], df['length_of_service'], alpha = 0.2, color = 'g')
plt.title('Scatterplot of Age vs Length of Service')
plt.xlabel('Age in years')
plt.ylabel('Length of Service in years')
plt.show()
````

#### 5.D Plot Histogram of store_name

````python
# A:
plt.hist(df['store_name'], bins = 50)
plt.title('Histogram of Store Names')
plt.xlabel('Store Name as numbers')
plt.ylabel('Frequency')
plt.show()

````
