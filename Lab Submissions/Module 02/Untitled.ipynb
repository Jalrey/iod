```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
```

```python
cars = pd.read_csv("C:/Users/jadea/Documents/IOD/02.1 - Module 02 Lab/Cars_kaggle.csv")
cars.head()
```

```python
cars.tail()
```

```python
cars.dtypes
```

```python
cars_droppedcol = cars.drop([ 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
cars_droppedcol.head()
```

```python
cars_droppedcol = cars_droppedcol.rename(columns = {"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price", "Engine_Fuel_Type":"EFT" })
cars_droppedcol.head()
```

```python
crs = cars_droppedcol
crs.head()
```

```python
duplicated_rows_crs = crs[crs.duplicated()]
print("number of duplicate rows:", duplicated_rows_crs.shape)
```

```python
crs.count()
```

```python
crs = crs.drop_duplicates()
crs.head()
crs.shape
crs.count()
```

```python
print(crs.isnull().sum())
```

```python
crs = crs.dropna()
crs.count()
```

```python
print(crs.isnull().sum())
```

```python
sns.boxplot(x = crs['Price'])
plt.show()
```

```python
sns.boxplot(x=crs['HP'])
```

```python
sns.boxplot(x = crs['Cylinders'])
```

```python
crs.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,5))
c= crs.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(crs['HP'], crs['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

```python
_group_by = ['Make']

_agg = {'MPG-H':'mean', 'MPG-C':'mean'}

crs_make_groupby = crs.groupby(by=_group_by).agg(_agg)
crs_make_groupby.sort_values('MPG-H', ascending = False).plot(kind = 'barh', figsize=(20,20))
```

```python
_group_by = ['Make']

_agg = {'Price':'mean'}

crs_price_groupby = crs.groupby(by=_group_by).agg(_agg)
crs_price_groupby.sort_values('Price', ascending = False)

crs_price_groupby.plot(kind='bar', figsize=(10,5))
plt.title("Average Price by make")
plt.ylabel('Average Price')
plt.xlabel('Make');
plt.show()
```

```python
crs.query("Make == 'Bugatti'")
#Looks like there's only 3 entries for the Bugatti which drives up the mean price
```

```python
_group_by = ['Transmission']

_agg = {'Price':'mean'}

crs_transmission_groupby = crs.groupby(by=_group_by).agg(_agg)
crs_transmission_groupby.sort_values('Price', ascending = False)

crs_transmission_groupby.plot(kind='barh', figsize=(10,5))
plt.title("Average Price by transmission")
plt.ylabel('Average Price')
plt.xlabel('Transmission');
plt.show()
```

```python
crs.query("Transmission == 'AUTOMATED_MANUAL'")
#Check the distribution of Automated Manual per Make and Price

crs_automanual = crs.query("Transmission == 'AUTOMATED_MANUAL'")

_group_by = ['Make']

_agg = {'Price':'mean'}

crs_price_groupby_Automanual = crs_automanual.groupby(by=_group_by).agg(_agg)
crs_price_groupby_Automanual.sort_values('Price', ascending = False).plot(kind='barh', figsize=(10,5))

plt.title("Average Price by make for all Automated_Manual")
plt.ylabel('Average Price')
plt.xlabel('Make');
plt.show()

#Buggatti might be dragging the price up of Automated_Manual
```

```python
_group_by = ['EFT']

_agg = {'MPG-H':'mean',
       'MPG-C':'mean'}

crs_eft_groupby = crs.groupby(by=_group_by).agg(_agg)
crs_eft_groupby.sort_values('MPG-H', ascending = False)

crs_eft_groupby.plot(kind='barh', figsize=(10,5))
plt.title("Average MPGH by FUEL Type")
plt.ylabel('FUEL TYPE')
plt.xlabel('Transmission');
plt.show()
```
