# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# get matplot to show charts inline
%matplotlib inline
````

````python
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

**1. Find Dimension of Array**

Find dimention of each array: height, weight, population
```
# Expected Ouput
np.ndim(height): 1
np.ndim(weight): 1
np.ndim(population): 2
```

````python
print(f"The dimension of the height array is: {np.ndim(height)}")
print(f"The dimension of the weight array is: {np.ndim(weight)}")
print(f"The dimension of the population array is: {np.ndim(population)}")
````

**8. Create a histogram of height and weight**

````python
plt.hist(height, bins = 30)
plt.xlabel("Height in m")
plt.ylabel("Frequency")
plt.title("Histogram of Height")
plt.show()


plt.hist(weight, bins = 30)
plt.xlabel("Weight in kg")
plt.ylabel("Frequency")
plt.title("Histogram of Weight")
plt.show()
````

**9. Create a boxplot of height and weight**

````python
plt.figure(figsize = (16,12))
plt.boxplot(height)
plt.xlabel("Height in m")
plt.ylabel("Frequency")
plt.title("Boxplot of Height")

plt.show()

plt.figure(figsize = (16,12))
plt.boxplot(weight)
plt.xlabel("Weight in kg")
plt.ylabel("Frequency")
plt.title("Boxplot of Weight")
plt.show()
````

**10. Create scatterplot of height and weight**

````python

plt.figure(figsize = (16,12))
plt.scatter(height, weight, color = 'r', alpha = 0.2)
plt.xlabel("Height in meters")
plt.ylabel("Weight in kilograms")
plt.title("Scatter Plot of Height vs Weight")
````
